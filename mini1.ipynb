```python
# 1. minirpoject as beginner a random no. generator game
import random
target=random.randint(1,100)
while True:
  userchoice=input("enter your choice (1-100)or QUIT THE GAME(press Q):  ")
  if userchoice =="Q":
    break
  userchoice=int(userchoice)
  if (userchoice==target):
    print("😁👍success: Correct guess!!  ")
    break
  if userchoice<target:
    print("😉take bigger guess")
  if userchoice>target:
    print("😒take a small guess")
print("GAME OVER ")

```
